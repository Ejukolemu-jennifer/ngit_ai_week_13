```python
from google.colab import userdata
from groq import Groq
import os

# Securely fetch the key from Colab Secrets
os.environ["GROQ_API_KEY"] = userdata.get("GROQ_API_KEY")

# Initialize the client
client = Groq()
print("✅ Client successfully initialized from Colab Secrets!")
```

```python
def generate_llm_response(prompt):
      completion = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}],
                            temperature=0.2
                                  )
      return completion.choices[0].message.content
```

```python
policy_text = """
Data Privacy and Security Policy (v2.1): All employees must use multi-factor authentication (MFA)
to access company networks. Passwords must be updated every 90 days and contain at least 12 characters,
including one uppercase letter, one number, and one special character. No customer data may be stored
on local drives or personal devices. Violations of this policy will lead to immediate disciplinary action,
up to and including termination. Remote workers must log in via the approved corporate VPN at all times.
"""
print("✅ Policy text loaded!")
```

```python
# Create the prompt for summarization
summary_prompt = f"Summarize the following policy text into three concise bullet points:\n\n{policy_text}"

# Call the LLM function
summary_result = generate_llm_response(summary_prompt)

# Print the result
print(summary_result)
```

```python
# Create the prompt for question answering
qa_prompt = f"Based on this policy text, what happens if an employee stores customer data on a local drive?\n\nContext:\n{policy_text}"

# Call the LLM function
qa_result = generate_llm_response(qa_prompt)

# Print the result
print(qa_result)
```

```python
# Format the final string containing both experiment results
final_output = f"""==================================================
LLM EXPERIMENT RESULTS
==================================================

TASK 1: SUMMARIZATION
--------------------------------------------------
Prompt Sent: {summary_prompt}

Model Response:
{summary_result}

==================================================
TASK 2: QUESTION ANSWERING
--------------------------------------------------
Prompt Sent: {qa_prompt}

Model Response:
{qa_result}
"""

# Open a file in write mode and save the content
with open("experiment_output.txt", "w", encoding="utf-8") as f:
    f.write(final_output)

print("✅ Results successfully saved to 'experiment_output.txt'!")
```
